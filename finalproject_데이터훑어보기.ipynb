```python
import pandas as pd

# CSV 파일 경로를 지정합니다.
csv_file = "/content/Chicago_Crimes_2001_to_2004.csv"  # 파일명을 실제 파일명으로 변경하세요

# 잘못된 줄을 건너뛰면서 CSV 파일 읽기
df = pd.read_csv(csv_file, on_bad_lines='skip')

print("CSV 파일 로드 성공. 데이터프레임 shape:", df.shape)

# 'Primary Type', 'Description', 'Location Description' 컬럼의 유니크 값 확인
primary_type_unique = df["Primary Type"].unique()
description_unique = df["Description"].unique()
location_description_unique = df["Location Description"].unique()

print("Primary Type 유니크 값:", primary_type_unique)
print("Description 유니크 값:", description_unique)
print("Location Description 유니크 값:", location_description_unique)
```
